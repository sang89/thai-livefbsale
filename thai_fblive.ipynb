```python
import numpy as np
import pandas as pd

#open the dataset
path = '../datasets/thai_fb_sale/thai_sale.csv'
data = pd.read_csv(path)
data.head(5)

# drop uncesssary columns
to_drop = ['status_id', 'Column1', 'Column2', 'Column3', 'Column4']
data.drop(to_drop, axis=1, inplace=True)
video = data[data['status_type']=='video']
video['status_published'] = video['status_published'].apply(pd.to_datetime)

def removeStuff(num):
    if num<0:
        return np.nan
    return num

video['num_reactions'] = video['num_reactions'].apply(removeStuff)
video['num_shares'] = video['num_shares'].apply(removeStuff)
video['num_comments'] = video['num_comments'].apply(removeStuff)
video.dropna(inplace=True)
video.head(5)

```

```python
from fbprophet import Prophet

table = pd.DataFrame()
# Time threshold
time = pd.to_datetime('2017-07-01 00:00:00')
table['ds'] = video['status_published']
#table['ds'] = table.where(table>time)
table['y'] = video['num_reactions']

table.dropna(inplace=True)

# Create FB prophet
m = Prophet(growth = 'linear')
m.fit(table)

# Create future forecast
future = m.make_future_dataframe(periods=10)

forecast = m.predict(future)
```

```python
m.plot(forecast)
```
